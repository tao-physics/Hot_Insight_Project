```python
data_pred['2nd_ratio'] = data_pred['prob_2nd'] / (1 - data_pred['prob_1st'])
```

```python
data_pred.head()
```

```python
sn.distplot(data_pred['prob_2nd'], label='all_prob_2nd')
sn.distplot(data_pred[data_pred['Label_encoded'] == data_pred['label_2ndpred']]['prob_2nd'], label='acc_prob_2nd')
plt.legend()
```

```python
sn.distplot(data_pred['2nd_ratio'], label='all_ratio_2nd')
sn.distplot(data_pred[data_pred['Label_encoded'] == data_pred['label_2ndpred']]['2nd_ratio'], label='acc_ratio_2nd')
plt.legend()
```

```python
label_final_pred = []
threshold = 0.45
for k in range(len(data_pred)):
    temp = data_pred['label_pred'][k]
    if (data_pred['2nd_ratio'][k] > threshold) & (data_pred['Label_encoded'][k] == data_pred['label_2ndpred'][k]):
        temp = data_pred['label_2ndpred'][k]
    label_final_pred.append(temp)
    
data_pred['label_final_pred'] = label_final_pred
```

```python
data_pred.head()
```

```python
print(classification_report(data_pred['Label_encoded'], data_pred['label_final_pred']))
```

```python
cm = confusion_matrix(data_pred['Label_encoded'], data_pred['label_final_pred'])
```

```python
plt.figure(figsize=(10,10))
sn.heatmap(cm, cmap='Purples', annot=True, fmt='.0f')
```

```python
plt.figure(figsize=(10,10))
sn.heatmap(cm / cm.sum(axis=1).reshape(len(cm), -1), cmap='Purples', annot=True, fmt='.2f')
```
